# Analysis of Euro Daily Exchange Rates (1999-2021)

In this project, we will explore a dataset that captures the daily exchange rates of the euro (€) against various global currencies from 1999 to 2021. The euro, being the official currency in most European Union countries, plays a crucial role in international trade and finance.

For example, if the exchange rate between the euro and the US dollar is 1.5, one euro would be worth 1.5 US dollars, indicating a higher value for the euro relative to the dollar at this rate.

The dataset, curated by Daria Chemkaeva and available on Kaggle, is sourced from the European Central Bank (ECB). Although regularly updated, our analysis is based on the version downloaded in January 2021.

This notebook will delve into trends and patterns in the euro's exchange rates, offering insights into its performance over the years.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline
exchange_rates = pd.read_csv('Dataset/euro-daily-hist_1999_2022.csv')
```

The DataFrame contains 6,535 rows and 41 columns. It represents the daily exchange rates of the euro against various global currencies. Some columns have missing values, such as the Bulgarian lev and Brazilian real, while others are fully complete, like the US dollar and Swiss franc.

### Data Cleaning

Our focus in this project will be on the exchange rate between the euro and the American dollar

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
```

```python
euro_to_dollar = euro_to_dollar[~euro_to_dollar['US_dollar'].str.contains('-', na=False)]
```

```python
euro_to_dollar.loc[:, 'US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.loc[:, "rolling_mean"] = euro_to_dollar["US_dollar"].rolling(30).mean()
```

### Rolling Mean

```python
euro_to_dollar['rolling_mean'][30:]
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

### Storytelling Data Visualization

We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

```python
bush_obama_trump = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy(
       )[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

Here, we used matplotlib's functional approach to create the graphs, as it provides greater flexibility in arranging subplots:

First, we arrange three graphs on a 2-by-3 grid (although this grid has space for six subplots, only three are created, leaving the bottom row empty). Next, we place one graph on the bottom of a 2-by-1 grid (with room for two subplots, but the top row remains empty). The two grids are combined, resulting in three graphs on the top row and one graph on the bottom row.

```python
import matplotlib.dates as mdates

style.use('fivethirtyeight')

plt.figure(figsize=(10, 5), dpi=100)
ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 2)
ax3 = plt.subplot(2, 3, 3)
ax4 = plt.subplot(2, 1, 2)
axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'], alpha=0.3)
    ax.grid(alpha=0.5)

def set_x_ticks(ax, start_year, end_year):
    years = mdates.YearLocator(2)
    years_fmt = mdates.DateFormatter('%Y')
    ax.xaxis.set_major_locator(years)
    ax.xaxis.set_major_formatter(years_fmt)
    ax.set_xlim(pd.Timestamp(f"{start_year}-01-01"), pd.Timestamp(f"{end_year}-12-31"))

# Ax1: Bush
ax1.plot(bush['Time'], bush['rolling_mean'], color='#0000FF')
set_x_ticks(ax1, 2001, 2009)
ax1.text(pd.Timestamp("2005-01-01"), 1.92, 'BUSH', fontsize=18, weight='bold', color='#0000FF')
ax1.text(pd.Timestamp("2004-01-01"), 1.8, '(2001-2009)', weight='bold', alpha=0.3)

# Ax2: Obama
ax2.plot(obama['Time'], obama['rolling_mean'], color='#FF0000')
set_x_ticks(ax2, 2009, 2017)
ax2.text(pd.Timestamp("2013-01-01"), 1.92, 'OBAMA', fontsize=18, weight='bold', color='#FF0000')
ax2.text(pd.Timestamp("2012-01-01"), 1.8, '(2009-2017)', weight='bold', alpha=0.3)

# Ax3: Trump
ax3.plot(trump['Time'], trump['rolling_mean'], color='#90EE90')
set_x_ticks(ax3, 2017, 2021)
ax3.text(pd.Timestamp("2019-01-01"), 1.92, 'TRUMP', fontsize=18, weight='bold', color='#90EE90')
ax3.text(pd.Timestamp("2018-06-01"), 1.8, '(2017-2021)', weight='bold', alpha=0.3)

# Ax4: Bush-Obama-Trump
ax4.plot(bush['Time'], bush['rolling_mean'], color='#0000FF')
ax4.plot(obama['Time'], obama['rolling_mean'], color='#FF0000')
ax4.plot(trump['Time'], trump['rolling_mean'], color='#90EE90')
ax4.grid(alpha=0.5)
set_x_ticks(ax4, 2001, 2021)

# Adding a title and a subtitle
plt.suptitle('EURO-USD rate averaged 1.22 under the last three US presidents', fontsize=20, weight='bold', y=1.05)

# Adjust layout and display the plot
plt.tight_layout()
plt.subplots_adjust(top=0.85, bottom=0.1)
plt.show()
```
